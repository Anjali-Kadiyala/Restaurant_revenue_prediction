```python
import pandas as pd
```

```python
data=pd.read_csv('train.csv')
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Open_Date = LabelEncoder()
data['Open_Date'] = Open_Date.fit_transform(data['Open_Date'])

City = LabelEncoder()
data['City'] = City.fit_transform(data['City'])

City_Group = LabelEncoder()
data['City_Group'] =City_Group.fit_transform(data['City_Group'])

Type = LabelEncoder()
data['Type'] = Type.fit_transform(data['Type'])

data.head()
```

```python
x=data.drop(['Id','Open_Date','revenue'],axis=1)
y=data['revenue']
x.head()
```

```python
data.isnull().sum()
```

## Tree Regressor

```python
from sklearn import tree
```

```python
rtree=tree.DecisionTreeRegressor()
```

```python
rtree.fit(x,y)
```

```python
feature_imp=pd.DataFrame()
feature_imp['col']=x.columns
feature_imp['imp']=rtree.feature_importances_*100

feature_imp.sort_values('imp',ascending=False)
```

```python
x=x.drop(['P18','P33','P32','P30','P13','P7','P36','P24','P35','City_Group','P31','Type','P12','P17','P14','P5','P27','P34','P11','P37','P16','P10'],axis=1)
x.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn import tree
import numpy as np
```

```python
rtree_model=rtree.fit(x, y)
y_pred=rtree.predict(x)
rtree
```

```python
y_train_pred= rtree_model.predict(x)
print(np.sqrt(metrics.mean_squared_error(y,y_train_pred)))
```

```python
y.mean()
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=.4)
```

```python
import numpy as np
rtree.fit(x_train, y_train)
y_test_pred=rtree.predict(x_test)
rtree

np.sqrt(metrics.mean_squared_error(y_test,y_test_pred))
```

```python
comp_df=x_test.copy()
comp_df['actual_value']=y_test
comp_df['predicted_value']=y_test_pred
comp_df.head()
```

```python
comp_df['predicted_value']=comp_df['predicted_value'].astype(np.int64)
```

```python
param_grid= {"criterion"          : ['mse', 'mae'],
             "max_features"       : [11,12,13,14,15],
             "splitter"           : ['best', 'random'],
             "min_samples_leaf"  : [5,6,7,8,9,10,11,12,13],
             "max_depth"          : [9,10,11,12,13,14]
            }
```

```python
optimized_tree=GridSearchCV(rtree, param_grid, scoring='neg_mean_absolute_error', cv=5)
```

```python
optimized_tree.fit(x_train,y_train)
```

```python
import sklearn
sklearn.metrics.SCORERS.keys()
```

```python
optimized_tree.best_params_
```

```python
y_test_predict=rtree_final.predict(x_test)
print(np.sqrt(metrics.mean_squared_error(y_test,y_test_predict)))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=1000, oob_score = 'TRUE',n_jobs = -1,random_state =50 ) 
rf.fit(x, y)
```

```python
comp = pd.DataFrame()
comp['Columns'] = x.columns
comp['Importance'] = rf.feature_importances_*100
comp.sort_values('Importance', inplace=True, ascending=False)
comp
```

```python
x.drop(['P3','P26','P2','P9'],axis=1)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.3)
print(x_train.shape, y_train.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=1000, oob_score = 'TRUE',n_jobs = -1,random_state =50 ) 
rf.fit(x_train, y_train)
y_test_pred = rf.predict(x_test)
print(np.sqrt(metrics.mean_squared_error(y_test,y_test_pred)))
```

```python
param_grid = { 'n_estimators':[100,500,1000],
              'oob_score': ['TRUE','FALSE'],
              'n_jobs':  [-1,1],
              'random_state': [5,10,40,50]}
```

```python
optimized_forest = GridSearchCV(rf,param_grid,scoring= 'neg_mean_absolute_error',verbose = 2, cv=5)
```

```python
optimized_forest.fit(x_train,y_train)
```

```python
optimized_forest.best_score_
```

```python
optimized_forest.best_params_
```

```python
optimized_forest.best_estimator_
```

```python
optimized_forest.best_index_
```

```python
final_model = optimized_forest.best_estimator_
```

```python
y_test_pred1 = final_model.predict(x_test)
```

```python
c = x_test.copy()
c['Actual'] = y_test
c['Prediction'] = y_test_pred1
c.head()
```

```python
print(np.sqrt(metrics.mean_squared_error(y_test,y_test_pred1)))
```
